```
import numpy as np
import pandas as pd
import seaborn as sns
```

```
df = sns.load_dataset('mpg')
```

```
df.head(3)
```

```
df.model_year.value_counts().sort_index(ascending = True)
```

```
df['bin_year'] = df.model_year.apply(lambda x: 'Vintage' if x<75 else 'Retro' if x<80 else 'Modern')
```

```
sns.relplot(
    data = df, x = 'mpg', y = 'acceleration', 
    col = 'origin',
    row = 'bin_year',
    hue = 'cylinders',
    palette = 'YlGnBu',
    kind = 'scatter'
    );
```

```
sns.relplot(data = df, x = 'model_year', y = 'mpg', hue = 'origin', kind = 'line', palette = 'YlGnBu')
```

```
sns.relplot(data = df, x = 'model_year', y = 'mpg', hue = 'origin', kind = 'scatter', palette = 'YlGnBu', legend = 'auto')
```

```
sns.scatterplot(data = df, x = 'mpg', y = 'weight', 
    hue = 'cylinders', size = 'mpg', style = 'origin', 
    legend = 'auto', palette = 'YlGnBu')
```

```
df.head(3)
```

```
df.eval('mpg/weight').rename('mileage per pound of weight')
```

```
df['mpg/weight'] = df.eval('mpg/weight').rename('mileage per pound of weight')
```

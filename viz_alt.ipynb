# Altair

```python
import numpy as np
import pandas as pd
import altair as alt
```

```python
csv_url = 'https://raw.githubusercontent.com/n2cholas/dsc-workshops/master/Intro_to_Interactive_Data_Viz_with_Altair/pokemon-data-cleaned.csv'
df = pd.read_csv(csv_url, na_filter=False)
df.head(3)
```

```python
df.drop(['Num Types', 'Type 1', 'Type 2', 'Num Abilities',
        'Ability 1', 'Ability 2', 'Ability 3', 'Has Negative Ability', 'HP',
        'Attack', 'Defense', 'Special Attack', 'Special Defense', 'Speed',
        'Next Evolution(s)', 'Evolutionary Stage',
        'Num Evolutionary Stages', 'Evolutionary Progress', 'Is Mega Evolution',
        'Is Alternate Form', 'Moves', 'Defensive Boost Moves',
        'Offensive Boost Moves', 'Max Defensive Boost Amount',
        'Max Offensive Boost Amount', 'Recovery Moves', 'Priority STAB Attacks',
        'Entry Hazards', 'Hazard Clearing Moves', 'Phazing Moves',
        'Switch Attacks', 'High Prob Side FX Attacks',
        'Constant Damage Attacks', 'Trapping Moves'], axis=1, inplace=True)
```

```python
alt.Chart(df).mark_point().encode(
    x = 'Base Stat Total'
)
```

```python
import seaborn as sns
cars = sns.load_dataset('mpg')
```

```python
alt.Chart(cars).mark_point().encode(
    x = 'mpg'
)
```

```python
alt.Chart(cars).mark_point().encode(
    x = 'mpg',
    y = 'origin'
)
```

```python
alt.Chart(cars).mark_tick().encode(
    x = 'mpg',
    y = 'origin'
).interactive()
```

```python
interval = alt.selection_interval()

alt.Chart(cars).mark_point().encode(
    x = 'mpg:Q',
    y = 'weight:Q',
    color = 'model_year'
).properties(
        selection = interval
    )
```

```python
interval = alt.selection_interval()

alt.Chart(cars).mark_point().encode(
    x = 'mpg:Q',
    y = 'weight:Q',
    color = alt.condition(interval, 'model_year', alt.value('lightgray'))
).properties(
        selection = interval
    )
```

```python
select = alt.selection_interval(encodings = ['y'])

alt.Chart(cars).mark_point().encode(
    x = 'weight',
    y = 'mpg',
    color = alt.condition(select, 'origin', alt.value('Lightgray'))
).properties(
        selection = select
    )
```

```python
select = alt.selection_interval(encodings = ['x', 'y'])

alt.Chart(cars).mark_point().encode(
    x = 'weight',
    y = 'mpg',
    color = alt.condition(select, 'origin', alt.value('Lightgray'))
).properties(
        selection = select
    )
```

```python
selection = alt.selection_interval(encodings = ['x', 'y'])

scatter = alt.Chart(cars).mark_point().encode(
    x = 'weight',
    y = 'mpg',
    color = alt.condition(selection, 'origin', alt.value('Lightgray'))
).properties(selection = selection)

bar = alt.Chart(cars).mark_bar().encode(
    x = 'mean(weight)',
    y = alt.Y('mpg', bin = alt.Bin(maxbins = 15)),
    color = 'origin'
)

scatter & bar
```

```python
selection = alt.selection_interval(encodings = ['x', 'y'])

scatter = alt.Chart(cars).mark_point().encode(
    x = 'weight',
    y = 'mpg',
    color = alt.condition(selection, 'origin', alt.value('Lightgray'))
).properties(selection = selection)

bar = alt.Chart(cars).mark_bar().encode(
    x = 'mean(weight)',
    y = alt.Y('mpg', bin = alt.Bin(maxbins = 15)),
    color = 'origin'
)

scatter | bar
```

```python
selection = alt.selection_interval(encodings = ['x', 'y'])

scatter = alt.Chart(cars).mark_point().encode(
    x = 'weight',
    y = 'mpg',
    color = alt.condition(selection, 'origin', alt.value('Lightgray'))
).properties(selection = selection)

bar = alt.Chart(cars).mark_bar().encode(
    x = 'mean(weight)',
    y = alt.Y('mpg', bin = alt.Bin(maxbins = 15)),
    color = 'origin'
)

alt.hconcat(scatter, bar)
```

```python
selection = alt.selection_interval(encodings = ['x', 'y'])

scatter = alt.Chart(cars).mark_point().encode(
    x = 'weight',
    y = 'mpg',
    color = alt.condition(selection, 'origin', alt.value('Lightgray'))
).properties(selection = selection)

bar = alt.Chart(cars).mark_bar().encode(
    x = 'mean(weight)',
    y = alt.Y('mpg', bin = alt.Bin(maxbins = 15)),
    color = 'origin'
).transform_filter(selection)

alt.vconcat(scatter, bar)
```

```python
cars_4_6_8 = cars.query('cylinders > 3')

multi = alt.selection_multi()

scatter = alt.Chart(cars_4_6_8).mark_point().encode(
    x = 'weight',
    y = 'mpg',
    color = 'cylinders:O'
)

bar = alt.Chart(cars_4_6_8).mark_bar().encode(
    x = alt.X('weight', bin = True),
    y = 'mean(mpg)'
)

scatter & bar
```

```python
select = alt.selection_multi(fields = ['origin'])

scatter = alt.Chart(cars).mark_point().encode(
    x = 'weight',
    y = 'mpg',
    color = alt.condition(select, 'origin', alt.value('Lightgray'))
).transform_filter(select)

bar = alt.Chart(cars).mark_bar().encode(
    x = 'mean(mpg)',
    y = 'cylinders:O',
    color = alt.condition(select, 'origin', alt.value('Lightgray'))
).properties(selection = select, title = 'mpg vs cylinders')

alt.vconcat(scatter, bar)
```

```python
select = alt.selection_multi(fields = ['origin'])
select_int = alt.selection_interval(encodings = ['y'])

scatter = alt.Chart(cars).mark_point().encode(
    x = 'weight',
    y = 'mpg',
    color = alt.condition(select & select_int, 'origin', alt.value('Lightgray'))
).properties(selection = select_int, title = 'mpy by weight')

bar = alt.Chart(cars).mark_bar().encode(
    x = 'mean(mpg)',
    y = 'cylinders:O',
    color = alt.condition(select, 'origin', alt.value('Lightgray'))
).properties(selection = select, title = 'avg. mpg by cylinders').transform_filter(select)

alt.vconcat(scatter, bar)
```

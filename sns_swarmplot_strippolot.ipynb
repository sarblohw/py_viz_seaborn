```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna(how = 'any')
```

```python
cars.cylinders.value_counts()
```

```python
cars.head()
```

```python
cars.dtypes
```

```python
cars.info()
```

```python
filt = (cars['cylinders'] == 4) | (cars['cylinders'] == 6) | (cars['cylinders'] == 8)
cars = cars[filt]
```

```python
cars[:5]
```

```python
sns.swarmplot(data = cars, x = 'mpg')
```

```python
sns.swarmplot(data = cars, x = 'origin', y = 'mpg')
```

```python
sns.swarmplot(cars.horsepower)
```

```python
sns.swarmplot(data = cars, x = 'origin', y = 'horsepower')
```

```python
sns.swarmplot(data = cars, x = 'origin', y = 'horsepower', hue = 'cylinders')
```

```python
sns.swarmplot(data = cars, x = 'origin', y = 'horsepower', hue = 'cylinders', dodge = True)
```

```python
#similarly we can use a swarmplot
sns.stripplot(data = cars, x = 'origin', y = 'horsepower', hue = 'cylinders', dodge = True)
```

```python
usa = cars.loc[cars.origin == 'usa']
```

```python
usa.head()
```

```python
#Overlaying swarm and boc plots
```

```python
sns.boxplot(y = usa.displacement, color='indigo')
sns.swarmplot(y = usa.displacement)
```

```python
sns.boxplot(x = usa.cylinders, y = usa.horsepower)
```

```python
sns.boxplot(x = usa.cylinders, y = usa.horsepower)
sns.swarmplot(x = usa.cylinders, y = usa.horsepower)
```

```python
sns.boxplot(x = usa.cylinders, y = usa.horsepower, whis = np.inf)
sns.swarmplot(x = usa.cylinders, y = usa.horsepower, color = 'black', alpha = 0.4)
```

```python
#combining vioilin and swarm
```

```python
sns.violinplot(x = usa.cylinders, y = usa.horsepower)
sns.swarmplot(x = usa.cylinders, y = usa.horsepower)
```

```python
sns.violinplot(x = usa.cylinders, y = usa.horsepower, scale = 'width')
sns.swarmplot(x = usa.cylinders, y = usa.horsepower, color = 'white')
```

```python
 sns.violinplot(x = usa.cylinders, y = usa.horsepower, scale = 'width')
sns.swarmplot(x = usa.cylinders, y = usa.horsepower, color = 'white', size = 4)
```

```python
 sns.violinplot(x = usa.cylinders, y = usa.horsepower, scale = 'width', inner = 'quartiles')
sns.swarmplot(x = usa.cylinders, y = usa.horsepower, color = 'white', size = 4)
```

```python
 sns.violinplot(x = usa.cylinders, y = usa.horsepower, scale = 'width', inner = None)
sns.swarmplot(x = usa.cylinders, y = usa.horsepower, color = 'white', size = 2.5)
```

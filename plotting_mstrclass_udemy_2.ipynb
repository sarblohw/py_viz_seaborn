# plt plotting

```
x = np.linspace(1, 10, 20)
y = np.random.randint(1, 50 ,20).cumsum()
```

```
x.size, y.size
```

```
plt.plot(y)
```

```
plt.plot(x, y)
```

```
plt.plot(x, y, color = 'r', marker = 'x', linestyle = '--')
```

```
plt.scatter(x, y, s = y, c = x, marker = 'x', linestyle = '--')
```

```
y2 = x*y

plt.subplot(1, 2, 1)
plt.plot(x, y, linestyle = '--')

plt.subplot(1, 2, 2)
plt.plot(x, y2, marker = 'o', linewidth = 2)
```

```
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.1, 0.6, 0.4, 0.3])

ax1.plot(x, y, c = 'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Yplt')

ax2.plot(x, y2, c = 'b')
ax2.set_xlabel('x')
ax2.set_ylabel('y2')
ax2.set_title('Y2plt')
```

```
fig, ax = plt.subplots(1, 2, figsize = (12, 4))

ax[0].plot(x, y)
ax[0].plot(x, y2)
ax[1].plot(x, y**2, c = 'k')

"""
ax[0].set_xlim([x, n])
ax[0].set_ylim([x, n])
ax[0].set_xticks([1, 4 ,9, 15])
ax[0].set_xticklabels([r'$\aplha\', '$\beta\'])
"""
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('data_10.csv')
```

```python
data.head(3)
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,6), layout='tight')
ax.plot(ages, dev_salaries, linestyle='--', linewidth=1, color='grey', label='all_devs')
ax.plot(ages, js_salaries, label='js_devs')
ax.plot(ages, py_salaries, label='py_devs', color='blue')
ax.fill_between(x=ages, y1=py_salaries, y2=js_salaries, where=(py_salaries>=js_salaries), interpolate=True, alpha=0.2)
ax.legend(loc='best')
ax.set_title('Salary by Age')
ax.set_xlabel('age')
ax.set_ylabel('salary usd')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10,6), layout='tight')

ax1.plot(ages, dev_salaries, linestyle='--', linewidth=1, color='grey', label='all_devs')

ax2.plot(ages, js_salaries, label='js_devs')
ax2.plot(ages, py_salaries, label='py_devs', color='blue')
ax2.fill_between(x=ages, y1=py_salaries, y2=js_salaries, where=(py_salaries>=js_salaries), interpolate=True, alpha=0.2)

ax1.legend(loc='best')
ax1.set_title('Salary by Age')
ax1.set_xlabel('age')
ax1.set_ylabel('salary usd')

ax2.legend(loc='best')
ax2.set_title('Salary by Age')
ax2.set_xlabel('age')
ax2.set_ylabel('salary usd')

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10,6), layout='tight', sharex=True)

ax1.plot(ages, dev_salaries, linestyle='--', linewidth=1, color='grey', label='all_devs')

ax2.plot(ages, js_salaries, label='js_devs')
ax2.plot(ages, py_salaries, label='py_devs', color='blue')
ax2.fill_between(x=ages, y1=py_salaries, y2=js_salaries, where=(py_salaries>=js_salaries), interpolate=True, alpha=0.2)

ax1.legend(loc='best')
ax1.set_title('Salary by Age')
ax1.set_ylabel('salary usd')

ax2.legend(loc='best')
ax2.set_xlabel('age')
ax2.set_ylabel('salary usd')

plt.show()
```

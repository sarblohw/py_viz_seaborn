```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')
```

```python
iris.plot()
```

```python
sns.heatmap(iris.corr(), annot=True)
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='day')
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='day', size='size', palette='YlGnBu')
```

```python
sns.histplot(data=tips.tip, kde=True, bins=15)
```

```python
plt.style.use('seaborn-pastel')
plt.hist(tips.tip, bins=15, label='Tips', density=True, orientation='vertical')
plt.xlabel('Tip Bins')
plt.ylabel('Frequency')
plt.title('Frequency Distribution')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7,4))
ax.hist(tips.tip, bins=15, orientation='vertical', density=True)
ax.set_xlabel('Tip Bins')
ax.set_ylabel('Frequency')
ax.set_title('Frequency Dist')
plt.show()
```

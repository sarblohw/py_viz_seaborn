```python
data_n = np.random.normal(loc=2500,size=100)
```

```python
plt.hist(data_n, bins=10, edgecolor='black', histtype='bar', log=True)
plt.xlabel('data')
plt.ylabel('freq_dist')
plt.rc('font', size=10, weight='bold')
plt.show()
```

```python
title_fonts = {'size': 10, 'weight': 'bold'}
label_fonts = {'size': 8}

plt.hist(data_n, bins=10, color='blue', alpha=0.2, edgecolor='black', label='nd_data', histtype='bar')
plt.title('Normal Distribution', fontdict=title_fonts)
plt.xlabel('bins', fontdict=label_fonts)
plt.ylabel('freq dist / probability', fontdict=label_fonts)
plt.xticks(ticks=[2470,2500,2550], labels=['a', 'b', 'c'])
plt.show()
```

```python
title_fonts = {'size': 10, 'weight': 'bold'}
label_fonts = {'size': 8}

plt.hist(data_n, bins=10, color='blue', alpha=0.2, edgecolor='black', label='nd_data', histtype='bar')
plt.title('Normal Distribution', fontdict=title_fonts)
plt.axvline(np.median(data_n))
plt.axhline(10)
plt.xlabel('bins', fontdict=label_fonts)
plt.ylabel('freq dist / probability', fontdict=label_fonts)
plt.show()
```

```python
title_fonts = {'size': 10, 'weight': 'bold'}
label_fonts = {'size': 8}

plt.hist(data_n, bins=10, color='blue', alpha=0.2, edgecolor='black', label='nd_data', histtype='bar')
plt.title('Normal Distribution', fontdict=title_fonts)
plt.axvline(np.median(data_n), color='aqua', linestyle='--', linewidth=1, label='median line')
plt.xlabel('bins', fontdict=label_fonts)
plt.ylabel('freq dist / probability', fontdict=label_fonts)
plt.legend(loc='best')
plt.show()
```

```
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```
tips = sns.load_dataset('tips')
```

```
tips.head(3)
```

```
sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'time', hue = 'day', style = 'smoker', size = 'size', palette = 'YlGnBu')
```

```
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    palette = 'YlGnBu',
    facet_kws=dict(sharex=False)
);
```

```
fmri = sns.load_dataset('fmri')
```

```
fmri.head(3)
```

```
sns.relplot(kind = 'line', 
    data = fmri, x = 'timepoint', y = 'signal', 
    col = 'region', hue = 'event', style = 'event', 
    palette = 'YlGnBu')
```

```
sns.lmplot(data = tips, x = 'total_bill', y = 'tip', col = 'sex', 
    hue = 'smoker', palette = 'YlGnBu', 
    facet_kws = dict(sharex = False))
```

```
sns.displot(
    data = tips, 
    x = 'tip', 
    kind = 'hist', 
    kde = True, 
    stat = 'count',
    col = 'sex',
    hue = 'size',
    palette = 'YlGnBu',
    element = 'step'
)
```

```
fig, ax = plt.subplots(1, 1, dpi = 100, layout = 'tight')
sns.histplot(
    data = tips,
    x = 'tip',
    hue = 'sex',
    kde = True,
    bins = 10,
    palette = 'YlGnBu',
    ax = ax
)
ax.set_title('Histplot')
ax.set_xlabel('Tip $')
ax.set_ylabel('Count of Tips')
ax.legend(title = "SEX", labels = ['M', 'F'],
            loc = 1, fontsize = 'small', fancybox = True)
plt.show()
```

```
sns.displot(
    data = tips,
    x = 'tip',
    hue = 'smoker',
    col = 'sex',
    kind = 'hist',
    kde = True,
    palette = 'YlGnBu',
    rug = True
)
```

```
sns.displot(
    data = tips,
    x = 'tip',
    hue = 'smoker',
    col = 'sex',
    kind = 'ecdf',
    palette = 'YlGnBu'
)
```

```
sns.ecdfplot(data = tips, x = 'tip', hue = 'sex', palette = 'YlGnBu');
```

```
sns.catplot(
    data = tips,
    x = 'sex',
    y = 'tip',
    kind = 'swarm',
    hue = 'time',
    palette = 'YlGnBu'
);
```

```
sns.catplot(
    data = tips,
    x = 'sex',
    y = 'tip',
    kind = 'strip',
    hue = 'time',
    palette = 'YlGnBu'
);
```

```
sns.catplot(
    data = tips,
    x = 'sex',
    y = 'tip',
    kind = 'bar',
    hue = 'time',
    palette = 'YlGnBu'
);
```

```
sns.catplot(
    data = tips,
    x = 'sex',
    y = 'tip',
    kind = 'violin',
    hue = 'time',
    palette = 'YlGnBu'
);
```

```
sns.violinplot(
    data = tips,
    y = 'tip',
    x = 'sex',
    hue = 'time',
    palette = 'YlGnBu',
    split = True,
    inner = None
);
```

```
fig, ax = plt.subplots(1, 1, dpi = 100, layout = 'tight')
sns.violinplot(
    data = tips,
    x = 'sex',
    y = 'total_bill',
    palette = 'YlGnBu',
    ax = ax
)
sns.swarmplot(
    data = tips,
    x = 'sex',
    y = 'total_bill',
    palette = 'YlGnBu',
    hue = 'smoker',
    ax = ax,
    linewidth = 1,
    legend = True
)
plt.show()
```

```
fig, ax = plt.subplots(1, 1, dpi = 100, layout = 'tight')
sns.boxplot(
    data = tips,
    x = 'sex',
    y = 'total_bill',
    palette = 'YlGnBu',
    ax = ax
)
sns.swarmplot(
    data = tips,
    x = 'sex',
    y = 'total_bill',
    palette = 'YlGnBu',
    hue = 'smoker',
    ax = ax,
    linewidth = 1,
    legend = True
)
plt.show()
```

```
penguins = sns.load_dataset("penguins")
```

```
penguins.head(3)
```

```
sns.jointplot(
    data = penguins,
    x = 'flipper_length_mm',
    y = 'body_mass_g',
    hue = 'species',
    palette = 'YlGnBu'
)
```

```
sns.pairplot(
    data = penguins, hue = 'species'
)
```

```
g = sns.PairGrid(penguins, hue="species", corner=True)
g.map_lower(sns.kdeplot, hue=None, levels=5, color=".2")
g.map_lower(sns.scatterplot, marker="+")
g.map_diag(sns.histplot, element="step", linewidth=0, kde=True)
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61, .6))
```

```
sns.set_theme(style="ticks", font_scale=1.25)
g = sns.relplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g",
    palette="crest", marker="x", s=100,
)
g.set_axis_labels("Bill length (mm)", "Bill depth (mm)", labelpad=10)
g.legend.set_title("Body mass (g)")
g.figure.set_size_inches(6.5, 4.5)
g.ax.margins(.15)
g.despine(trim=True)
```

```
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack", element = 'poly')
```

```
cars = sns.load_dataset('mpg')
```

```
sns.lineplot(
    data = cars,
    x = 'model_year',
    y = 'mpg'
)
```

```
data = cars.groupby('model_year')['mpg'].agg('mean').reset_index()
```

```
data
```

```
sns.lineplot(
    data = data,
    x = 'model_year',
    y = 'mpg'
)
```

```
cars
```

```
cars\
    .loc[((cars.model_year >= 70) & (cars.model_year <= 75)) & ((cars.cylinders >= 4) & (cars.cylinders <= 8))]\
    .groupby(['origin', 'cylinders'])['mpg'].agg('mean').reset_index()
```

```
data = cars\
    .loc[((cars.model_year >= 70) & (cars.model_year <= 75)) & ((cars.cylinders >= 4) & (cars.cylinders <= 8))]\
    .groupby(['origin', 'cylinders', 'model_year'])['mpg'].agg('mean').reset_index()

sns.lineplot(
    data = data,
    x = 'model_year',
    y = 'mpg',
    hue = 'cylinders'
)
```

```
sns.displot(
    data = data,
    x = 'mpg',
    hue = 'cylinders',
    col = 'origin',
    rug = True,
    kde = True,
    element = 'step'
);
```

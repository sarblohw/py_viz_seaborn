```python
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
dates = [
    dt.datetime(2019, 5, 24),
    dt.datetime(2019, 5, 25),
    dt.datetime(2019, 5, 26),
    dt.datetime(2019, 5, 27),
    dt.datetime(2019, 5, 28),
    dt.datetime(2019, 5, 29),
    dt.datetime(2019, 5, 30)
]
```

```python
y = [0, 1, 3, 4, 6, 5, 7]
```

```python
plt.plot_date(dates, y)
plt.show()
```

```python
plt.plot_date(dates, y, linestyle='-', marker=None)
plt.show()
```

```python
plt.plot_date(dates, y, linestyle='-', marker=None)
plt.gcf().autofmt_xdate(rotation=45)
plt.gca().
plt.show()
```

```python
plt.plot_date(dates, y, linestyle='-', marker='o', linewidth=1.3, label='time series')
plt.gcf().autofmt_xdate(rotation=90)
plt.gca().xaxis.set_major_formatter(mpl.dates.DateFormatter('%b,%d,%Y'))
plt.rc('font', size=8)
plt.xlabel('time')
plt.ylabel('data')
plt.title('time series')
plt.tight_layout()
plt.grid(False)
plt.legend(loc='best')
plt.show()
```

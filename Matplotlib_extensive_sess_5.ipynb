```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head(3)
```

```python
df.LanguageWorkedWith.str.split(';').explode().value_counts().head(5)
```

```python
slices = [60, 40]
```

```python
plt.style.use('seaborn-pastel')
plt.pie(slices)
plt.title('pie chart')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('seaborn-pastel')
plt.pie(df.LanguageWorkedWith.str.split(';').explode().value_counts().head(5).values,
       labels = df.LanguageWorkedWith.str.split(';').explode().value_counts().head(5).index
        + ' ' +
       df.LanguageWorkedWith.str.split(';').explode().value_counts().head(5).values.astype('str'))
plt.title('Top 5 most popular languages')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('seaborn-pastel')
plt.pie(df.LanguageWorkedWith.str.split(';').explode().value_counts().head(5).values,
       labels = df.LanguageWorkedWith.str.split(';').explode().value_counts().head(5).index
        + ' ' +
       df.LanguageWorkedWith.str.split(';').explode().value_counts().head(5).values.astype('str'), 
        wedgeprops = {'edgecolor' : 'black'})
plt.title('Top 5 most popular languages')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('seaborn-pastel')
plt.pie([60, 30, 20], 
        labels=['a', 'b', 'c'], 
        colors=['#E0BBE4', '#D291BC', '#FFDFD3'], 
        wedgeprops={'edgecolor': 'grey'})
plt.title('pie chart')
plt.tight_layout()
plt.show()
```

```python
val = df.LanguageWorkedWith.str.split(';').explode().value_counts().head(5).values
lab = df.LanguageWorkedWith.str.split(';').explode().value_counts().head(5).index

plt.style.use('seaborn-dark')
plt.pie(val, labels=lab, wedgeprops={'edgecolor': 'grey'})
plt.title('Top 5 most popular languages')
plt.tight_layout()
plt.show()
```

```python
list(df.LanguageWorkedWith.str.split(';').explode().value_counts().head(5).index).index('Python')
```

```python
exp = np.zeros(len(val), dtype='float')
exp[list(df.LanguageWorkedWith.str.split(';').explode()
                                     .value_counts().head(5).index).index('Python')] = 0.1
```

```python
exp
```

```python
plt.style.use('seaborn-dark')
plt.pie(val, labels=lab, wedgeprops={'edgecolor': 'grey'}, explode=exp)
plt.title('Top 5 most popular languages')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('seaborn-pastel')
plt.pie([60, 30, 20], 
        labels=['a', 'b', 'c'], 
        colors=['#E0BBE4', '#D291BC', '#FFDFD3'], 
        wedgeprops={'edgecolor': 'grey'},
       explode=[0,0,0.2])
plt.title('pie chart')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('seaborn-pastel')
plt.pie([60, 30, 20], 
        labels=['a', 'b', 'c'], 
        colors=['#E0BBE4', '#D291BC', '#FFDFD3'], 
        wedgeprops={'edgecolor': 'grey'},
       explode=[0,0,0.2],
       shadow=True, startangle=90, autopct='%1.1f%%')
plt.title('pie chart')
plt.tight_layout()
plt.show()
```

```python
x = [60, 30, 20]

plt.style.use('seaborn-pastel')
plt.pie(x, 
        labels=['a', 'b', 'c'], 
        colors=['#E0BBE4', '#D291BC', '#FFDFD3'], 
        wedgeprops={'edgecolor': 'grey'},
       explode=[0,0,0.2],
       shadow=True, startangle=90, 
        autopct=lambda p: f'{p:.1f}%')
plt.title('pie chart')
plt.tight_layout()
plt.show()
```

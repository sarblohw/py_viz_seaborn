```
class Dogs():

    species = 'mammal'

    def __init__(self, name, breed, spots):
        self.name = name
        self.breed = breed
        self.spots = spots

    def bark(self):
        print(f'Woof! my name is {self.name}. I am a {Dogs.species}.')
```

```
dog_2 = Dogs(name = 'sam', breed = 'human', spots = False)
```

```
dog_2.bark()
```

```
dog_2.species
```

```
class Animal():

    species = 'Mammal'

    def __init__(self):
        print('Animal Created')

    def who_am_i(self):
        print('I am an animal')

    def eat(self):
        print('I am eating')

class Dogs(Animal):

    species = 'sss'

    def __init__(self):
        Animal.__init__(self)
        print('Dog Created')

    def eat(self):
        print('I am a dog and eating')

    def bark(self):
        print('Woof')
```

```
x = Animal()
```

```
y = Dogs()
```

```
x.eat()
```

```
y.eat()
```

```
x.species
```

```
y.species
```

```python
import pandas as pd
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins[:3]
```

```python
penguins.info()
```

```python
penguins.dropna(how='any', inplace=True)
```

```python
penguins.shape
```

```python
sns.histplot(x = penguins.bill_length_mm, data = 'penguins')
```

```python
sns.histplot(x = penguins.bill_length_mm, data = 'penguins', bins = 20)
```

```python
sns.histplot(y = penguins.bill_length_mm, data = 'penguins', bins = 20)
```

```python
sns.histplot(x = penguins.bill_length_mm, data = 'penguins', bins = 10, kde = True)
```

```python
sns.histplot(x = penguins.bill_length_mm, data = 'penguins', bins = [30, 45, 50], kde = True)
```

```python
sns.histplot(x = penguins.bill_length_mm, data = 'penguins', binwidth = 10)
```

```python
sns.histplot(x = penguins.bill_length_mm, data = 'penguins', 
             binwidth = 5, binrange = (30, 60))
```

```python
sns.histplot(x = penguins.bill_length_mm, data = 'penguins', 
             bins=10, stat='count')
```

```python
sns.histplot(x = penguins.bill_length_mm, data = 'penguins', 
             bins=10, stat='density')
#stat can be changed to density in which case the area of the plot will sum up to be 1
#this ways we can plot the relative probability
#here, the height of the bar multiplied by its width and then adding this 
#for every bar will be 1
```

```python
sns.histplot(data = penguins, x = 'bill_length_mm', bins = 10, stat = 'probability')
#probability of a pengin ending up in a bar
#here the height of all the bars will sum up to one
```

```python
sns.histplot(data = penguins, x = 'bill_length_mm', stat = 'proportion')
```

```python
sns.histplot(data = penguins, x = 'bill_length_mm', stat = 'percent')
```

```python
sns.histplot(data = penguins, x = 'bill_length_mm', 
             bins = 10, stat = 'probability', 
            fill = False, element = 'step')
```

```python
sns.histplot(data = penguins, x = 'bill_length_mm', 
             bins = 10, stat = 'probability', 
            fill = False, element = 'step', cumulative = True)
```

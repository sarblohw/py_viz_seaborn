```python
interval = alt.selection_interval(encodings = ['y'])

scatter = alt.Chart(df).mark_point().encode(
    x = 'Base Stat Total',
    y = 'Num Moves',
    color = alt.condition(interval, 'Tier', alt.value('Lightgray'))
).properties(
    selection = interval
)

bar = alt.Chart(df).mark_bar().encode(
    x = 'count()',
    y = 'Tier',
    color = 'Tier'
).transform_filter(
    interval
)

scatter & bar
```

```python
sel = alt.selection_interval(encodings = ['x'])

scatter = alt.Chart(df).mark_point().encode(
    x = 'Base Stat Total',
    y = 'Num Moves',
    color = alt.condition(sel, 'Tier', alt.value('Lightgray'))
).properties(
    selection = sel
)

bar = alt.Chart(df).mark_bar().encode(
    x = 'count()',
    y = 'Tier',
    color = 'Tier'
).transform_filter(
    sel
)

scatter & bar
```

```python
multi = alt.selection_multi(fields = ['Tier'])

scatter = alt.Chart(df).mark_point().encode(
    x = 'Base Stat Total',
    y = 'Num Moves',
    color = alt.condition(multi, 'Tier', alt.value('Lightgray'))
).transform_filter(
    multi
)

bar = alt.Chart(df).mark_bar().encode(
    x = 'count()',
    y = 'Tier',
    color = alt.condition(multi, 'Tier', alt.value('Lightgray'))
).properties(
    selection = multi,
    title = 'BST vs Num Moves'
)

scatter & bar
```

```python
multi = alt.selection_multi(fields = ['Tier'])
intv = alt.selection_interval(encodings = ['y'], empty = 'all')

scatter = alt.Chart(df).mark_point().encode(
    x = 'Base Stat Total',
    y = 'Num Moves',
    color = alt.condition(multi | intv, 'Tier', alt.value('Lightgray'))
).transform_filter(
    multi
).properties(
    selection = intv
)

bar = alt.Chart(df).mark_bar().encode(
    x = 'count()',
    y = 'Tier',
    color = alt.condition(multi, 'Tier', alt.value('Lightgray'))
).properties(
    selection = multi,
    title = 'BST vs Num Moves'
).transform_filter(
    intv
)

scatter & bar
```

```python
selec = alt.selection_interval()

scatter = alt.Chart(df).mark_point().encode(
    x = 'Base Stat Total',
    y = 'Num Moves',
    color = alt.condition(selec, 'Tier', alt.value('Lightgray'))
).properties(
    selection = selec
).interactive()

scatter
```

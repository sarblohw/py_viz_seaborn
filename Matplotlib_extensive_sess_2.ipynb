```python
plt.xkcd()
plt.plot(ages, dev, label='All Devs')
plt.plot(ages, py_dev, label='Py Devs')
plt.title('Median Salary USD vs Age')
plt.xlabel('Age')
plt.ylabel('Median Salary USD')
plt.legend()
plt.show()
```

```python
plt.plot(ages, dev, 'k--',label='All Devs')
plt.plot(ages, py_dev, 'b',label='Py Devs')
plt.title('Median Salary USD vs Age')
plt.xlabel('Age')
plt.ylabel('Median Salary USD')
plt.legend()
plt.show()
```

```python
plt.plot(ages, dev, color='#444444', linestyle='--', linewidth='1', marker = 'x',label='All Devs')
plt.plot(ages, py_dev, color='#5a7d9a',label='Py Devs')
plt.title('Median Salary USD vs Age')
plt.xlabel('Age')
plt.ylabel('Median Salary USD')
plt.legend()
plt.show()
```

```python
plt.xkcd()
plt.style.use('ggplot')
plt.plot(ages, dev, color='#444444', linestyle='--', linewidth='1', marker = 'x',label='All Devs')
plt.plot(ages, py_dev, color='#5a7d9a',label='Py Devs')
plt.title('Median Salary USD vs Age')
plt.xlabel('Age')
plt.ylabel('Median Salary USD')
plt.legend()
plt.show()
```

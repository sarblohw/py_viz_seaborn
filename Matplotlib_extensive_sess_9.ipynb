```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('csdata.csv')
```

```python
df.head(3)
```

```python
plt.scatter(x=df.view_count, y=df.likes)
plt.show()
```

```python
plt.scatter(x=df.view_count, y=df.likes, marker='o', s=25, c='red', alpha=0.2, edgecolor='black')
plt.tight_layout()
plt.show()
```

```python
plt.scatter(x=df.view_count, y=df.likes, marker='o', s=40, c='red', alpha=0.2, edgecolor='black', linewidth=1.5)
plt.tight_layout()
plt.show()
```

```python
plt.scatter(x=df.view_count, 
    y=df.likes, 
    marker='o', 
    s=df.ratio, 
    c=df.ratio,
    cmap='summer', 
    alpha=0.2, 
    edgecolor='black', 
    linewidth=1.5)

plt.title('scatter')
plt.xlabel('views')
plt.ylabel('likes')

cbar = plt.colorbar()
cbar.set_label('ratio')

plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
plt.scatter(x=df.view_count, 
    y=df.likes, 
    marker='o', 
    s=df.ratio, 
    c=df.ratio,
    cmap='summer', 
    alpha=0.2, 
    edgecolor='black', 
    linewidth=1.5,
    label='scatter')

x_coord = np.linspace(10e1,10e7,100)
b, a = np.polyfit(df.view_count, df.likes, deg=1)

plt.plot(x_coord, a + (b*x_coord), linestyle='--', linewidth=1, color='grey')

plt.title('scatter')
plt.xlabel('views')
plt.ylabel('likes')

cbar = plt.colorbar()
cbar.set_label('ratio')

plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

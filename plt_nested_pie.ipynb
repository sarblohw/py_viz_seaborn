```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
vals = np.array([[60, 32], [35, 20], [26, 36]])
vals_sum = vals.sum(axis = 1)
vals_flat = vals.flatten()
```

```python
cmap = plt.get_cmap('tab20c')
outer_colors = cmap(np.arange(3)*12)
inner_colors = cmap(np.arange(3)*7)
```

```python
f = dict(family = 'serif', size = 14, weight = 'bold', color = 'darkred')

fig, ax = plt.subplots(dpi = 100)
ax.pie(vals_sum, radius = 1, colors = outer_colors, wedgeprops = dict(width = 0.3, edgecolor = 'w'), 
    labels = vals_sum, labeldistance = 0.82, textprops = {'fontsize' : 14, 'color' : 'w'})
pie2_patches, pie2_texts, pie2_autotexts = ax.pie(vals_flat, radius = 0.7, colors = inner_colors, wedgeprops = dict(width = 0.3, edgecolor = 'w'), 
    autopct = '%1.1f%%')
pie2_autotexts[0].set_backgroundcolor('black')
plt.title('Doughnut Chart/Nested Pie Chart', fontdict = f)
plt.text(-0.1, -0.07, "HB", fontdict = f)
plt.show()
```

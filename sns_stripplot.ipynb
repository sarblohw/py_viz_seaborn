### Stripplot

- Jitter
- normally, when plotting scatter, similar points will overlap each other in the plot
- Jitter is introduced to move the points to make em visible
- like the swarmplot, stripplot is a ctegorical scatter plot
- swarm and strip convey very similar info
- the diff b/w strip and swarm is how the points are split
- swarm: the points are arranged in columns so the width will depict the distribution
- strip: the points are randomly separated (jitter) so the dist func is not visible
- strip instead uses color intensity to depict density
- in strip more data can be depicted since itll take lesser space

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna(how = 'any')
```

```python
sns.stripplot(data = cars, x = 'mpg')
```

```python
sns.stripplot(data = cars.loc[(cars.cylinders.isin([4,6])) & (cars.origin.isin(['japan', 'usa']))], 
              y = 'mpg', x = 'cylinders', hue = 'origin', dodge = True)
```

```python
sns.stripplot(cars.weight, cars.origin)
```

```python
sns.stripplot(data = cars.loc[cars.cylinders.isin([4,6,8])], 
              x = 'weight', y = 'origin', hue = 'cylinders')
```

```python
sns.stripplot(data = cars.loc[cars.cylinders.isin([4,6,8])], 
              x = 'weight', y = 'origin', hue = 'cylinders', dodge = True)
```

```python
sns.stripplot(data = cars.loc[cars.cylinders.isin([4,6,8])], 
              y = 'weight', x = 'origin', hue = 'cylinders', dodge = True)
```

```python
sns.stripplot(data = cars.loc[cars.cylinders.isin([4,6,8])], 
              y = 'weight', x = 'origin', hue = 'cylinders', dodge = True, jitter = False)
```

```python
sns.stripplot(data = cars.loc[cars.cylinders.isin([4,6,8])], 
              y = 'weight', x = 'origin', hue = 'cylinders', dodge = True, jitter = 0.05)
```

```python
from ipywidgets import interactive

def jitter_widget(jitter = 0.1):
    sns.stripplot(data = cars.loc[cars.cylinders.isin([4,6,8])], 
                  x = 'mpg', y = 'origin', jitter = jitter)
    
interactive(jitter_widget, jitter = (0, 0.65, 0.05))
```

```python
sns.stripplot(data = cars.loc[cars.cylinders.isin([4,6,8])], 
              y = 'weight', x = 'origin', hue = 'cylinders', 
              dodge = True, jitter = 0.3, alpha = 0.3, size = 7, linewidth = 2)
```

```python
sns.stripplot(data = cars.loc[cars.cylinders.isin([4,6,8])], 
              y = 'weight', x = 'origin', hue = 'cylinders', 
              dodge = True, jitter = 0.3, alpha = 0.3, size = 7, 
              linewidth = 2, edgecolor = 'black', marker = '*')
```

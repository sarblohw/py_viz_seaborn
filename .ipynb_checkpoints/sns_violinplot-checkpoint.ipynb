```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

[https://www.youtube.com/watch?v=PNNLefP974M&list=PLtPIclEQf-3cG31dxSMZ8KTcDG7zYng1j&index=6]

- Violinplot is 1 part boxplot and 2 parts kdeplot

- categorical distribution plot

- box and violin plots show very similar summary stats

- median, IQR, whiskers

- box features outliers whereas violin will show a KDE of the data showing the probability density function

```python
cars = sns.load_dataset('mpg').dropna(how = 'any')
```

```python
cars.head(3)
```

```python
sns.violinplot(data = cars, x = 'origin', y = 'mpg')
```

```python
cars = cars.loc[cars.cylinders.isin([4, 6 ,8])]
```

```python
cars[:5]
```

```python
sns.violinplot(data = cars, x = 'displacement')
```

```python
sns.violinplot(data = cars, y = 'displacement')
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement)
```

```python
sns.violinplot(data = cars, x = 'cylinders', y = 'displacement', hue = 'origin')
```

```python
sns.violinplot(data = cars.loc[cars.origin.isin(['japan', 'europe'])],
                x = 'cylinders', y = 'displacement', hue = 'origin')
plt.legend(loc=2)
plt.show()
```

```python
sns.violinplot(data = cars.loc[cars.origin.isin(['japan', 'europe'])],
                x = 'cylinders', y = 'displacement', hue = 'origin', split = True)
plt.legend(loc=2, title = 'origin');
#for the split to work there has to be 2 catrgories in the hue parameter
```

```python
sns.violinplot(data = cars.loc[cars.origin.isin(['japan', 'europe'])],
                x = 'cylinders', y = 'displacement', hue = 'origin', split = True, inner = 'quartiles')
plt.legend(loc=2);
```

```python
sns.violinplot(data = cars.loc[cars.origin.isin(['japan', 'europe'])],
                x = 'cylinders', y = 'displacement', hue = 'origin', split = True, inner = 'quartiles', scale = 'count')
plt.legend(loc=2);
#scaling will have a diff scales for the 2 sides of the violin
#note that the plot is first filtering the data by 4 & 6 cylinders and then scaling by count
```

```python
sns.violinplot(data = cars.loc[cars.origin.isin(['japan', 'europe'])],
                x = 'cylinders', y = 'displacement', hue = 'origin', split = True, inner = 'quartiles', scale = 'count',
                scale_hue = False)
plt.legend(loc=2);
#here we can now see the overall number of values present in each group
```

```python
sns.violinplot(x = cars.origin, y = cars.displacement, order = ['japan', 'europe', 'usa'])
```

```python
sns.violinplot(x = cars.origin, y = cars.displacement, order = ['japan', 'europe', 'usa'], linewidth = 3)
```

```python
sns.violinplot(x = cars.origin, y = cars.displacement, order = ['japan', 'europe', 'usa'], linewidth = 3, bw = 0.2)
```

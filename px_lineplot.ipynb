```python
df_stocks = px.data.stocks()
df_stocks.sample(10)
```

```python
fig = px.line(data_frame = df_stocks, x = 'date', y = 'GOOG', 
              labels = dict(x = 'Date', y = 'Price'))
fig.show()
```

```python
fig = px.line(data_frame = df_stocks, x = 'date', y = ['AAPL', 'GOOG'], 
              labels = {'x': 'Date', 'y': 'Price'}, title = 'Apple vs Google Stock Price')
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = df_stocks.date, y = df_stocks.AAPL, mode = 'lines', 
                        name = 'Apple'))
fig.add_trace(go.Scatter(x = df_stocks.date, y = df_stocks.AMZN, mode = 'lines+markers', 
                        name = 'Amazon'))
fig.add_trace(go.Scatter(x = df_stocks.date, y = df_stocks.GOOG, mode = 'lines', 
                        name = 'Google', line = {'color': 'firebrick', 'width': 2,
                                                 'dash': 'dashdot'}))
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = df_stocks.date, y = df_stocks.AAPL, mode = 'lines', 
                        name = 'Apple'))
fig.add_trace(go.Scatter(x = df_stocks.date, y = df_stocks.AMZN, mode = 'lines+markers', 
                        name = 'Amazon'))
fig.add_trace(go.Scatter(x = df_stocks.date, y = df_stocks.GOOG, mode = 'lines', 
                        name = 'Google', line = {'color': 'firebrick', 'width': 2,
                                                 'dash': 'dashdot'}))
fig.update_layout(title = 'Stock Price Data 2018 - 2020', xaxis_title = 'Price', 
                 yaxis_title = 'Date')
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('survey_results_public.csv')
```

```python
df.LanguageWorkedWith.head(3)
```

```python
data = df.LanguageWorkedWith.str.split(';').explode().value_counts().sort_values(ascending=False)[:10]
```

```python
data
```

```python
pd.options.display.max_rows = 100
```

```python
data[::-1].plot(kind='barh')
```

```python
plt.style.use('seaborn-pastel')
plt.barh(data.index, data.values[::-1], label='sal by lang')
plt.title('median sal usd by language')
plt.xlabel('languages')
plt.ylabel('median_salary_USD')
plt.legend(loc='best')
plt.show()
```
